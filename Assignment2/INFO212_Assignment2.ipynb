#### Qestion 3 [20 marks]:
Given the following list of file names: 
```
filenames = ['inflammation-02.csv', 'inflammation-04.csv', 'inflammation-06.csv']
```
Write a program to plot the daily averages, maximums, and minimums of inflammation for each of the file. Your plot should have three rows each of which shows side-by-side plots of average, max, and min. The following figure illustrates the expected outcome:
![](https://i.imgur.com/XXT2Jqw.png)

#### Question 3 Answer

````python
# List of filenames
filenames = ['inflammation-02.csv', 'inflammation-04.csv', 'inflammation-06.csv']
````

````python
# Open each file from filenames as a numpy array and store in list.
file_arrays = [np.array(pd.read_csv(folder_dir + "/" + filenames[x])).astype(np.float32) for x in range(len(filenames))]
````

````python
# Function to plot len(file_list) Files on the same plot, with daily averages, maximums and minimums. The function takes in a list of numpy arrays.
def plot_files_with_averages(file_list):

    # Check if file_list is a list of numpy arrays.
    if not all(isinstance(x, np.ndarray) for x in file_list):
        raise TypeError("All elements in file_list must be numpy arrays.")
        return None

    # Change the font size of the plot.
    matplotlib.rcParams.update({'font.size': 16})

    
    ax, fig = plt.subplots(len(file_list), 3, figsize=(30, 20)) # Create a figure with 3 subplots.
    for row in range(len(file_list)): # For each array from file_list.
        for column in range(3): # For each different plot.
            if column == 0: # If column is 0, plot the daily averages.
                fig[row, column].plot(np.mean(file_list[row], axis=1)) # Plot the daily averages.
                metric = "Average" # Set the plot label to "Average".
            elif column == 1: # If column is 1, plot the maximums.
                fig[row, column].plot(np.max(file_list[row], axis=1)) # Plot the daily maximums. 
                metric = "Maximum" # Set the plot label to "Maximum".
            elif column == 2: # If column is 2, plot the minimums. 
                fig[row, column].plot(np.min(file_list[row], axis=1)) # Plot the daily minimums.
                metric = "Minimum" # Set the plot label to "Minimum".
            fig[row, column].set_title(metric + " Inflammation of " + filenames[row][:-4] + " Patients") # Set the title of each subplot.
            fig[row, column].set_xlabel("Patient") # Set the x-axis label of each subplot.
            fig[row, column].set_ylabel(metric + " Inflammation") # Set the y-axis label of each subplot.

    plt.tight_layout() # make sure the layout is more compact & readable. 
    plt.show() # Display the plot.
````

````python
plot_files_with_averages(file_arrays) # Call the function to plot the files.
````

3. What is the median of the price?

4. Create a new column 'price_norm' by subtracting the median price from the price column. Use the DataFrame.apply method.

````python
# Creating new 'price_norm' column in the dataframe. Engineering new column to be the 'price' - median_price, using df.apply method.
winemag['price_norm'] = winemag.apply(lambda row: row['price'] - median_price, axis=1) # Compute the price_norm column.
````
